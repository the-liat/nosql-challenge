### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}
result = list(esta.find(query))

# Use count_documents to display the number of documents in the result
print('-' * 40)
print("Number of documents in result:", esta.count_documents(query))
print('-' * 40)

# Display the first document in the results using pprint
pprint(result[0])
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
''' Since the Rating Value is stored as a string I need to first convert the field to integer
I used te following syntax originally - esta.update_many({}, [{"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}]), 
but it failed because there are some values that cannot be turned into numbers so I need to filter them out
(found the list of values from the error messages I got when trying to convert)'''

# creating the filter - which values to disregard
query = {'RatingValue': {'$nin': ['AwaitingInspection', 'Exempt', 
                                  'AwaitingPublication', 'Awaiting Inspection', 'Pass']}}
#creating the conversion to integer
update = [{"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}]

# updating the collection using the filter and the conversion statement
esta.update_many(query, update)

# checking that the field got updated (number not in quotes)
pprint(esta.find_one())
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex': 'London'}, 
         'RatingValue': {'$gte': 4}}
result = list(esta.find(query))

# Use count_documents to display the number of documents in the result
print('-' * 40)
print("Number of documents in result:", esta.count_documents(query))
print('-' * 40)

# Display the first document in the results using pprint
pprint(result[0])
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
print(f"Number of rows in the data frame: {london_df.shape[0]}")

# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
'''Now I realized that I didn't have to convert the rating field to integer. 
I could have just asked for ratings of 4 or 5 for the previous question...'''
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query = {"geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search},
         "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
         'RatingValue': 5,
         'scores.Hygiene': {"$ne": None}
        }
    
sort =  [('scores.Hygiene', -1)]

results = list(esta.find(query).sort(sort))

# Print the results
print('-' * 40)
print("Number of documents in result:", esta.count_documents(query))
print('-' * 40)
pprint(results)
```

```python
# Convert result to Pandas DataFrame
lat_lon_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f"Number of rows in the data frame: {lat_lon_df.shape[0]}")

# Display the first 10 rows of the DataFrame
lat_lon_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 

# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': '$LocalAuthorityName', 
                          'Number of Establishments': {'$sum': 1}
                         }
              }
# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'Number of Establishments': -1 }}

pipeline = [match_query, group_query, sort_values]
result = list(esta.aggregate(pipeline))

# Print the number of documents in the result
print('-' * 40)
print(f"Number of documents in the result: {len(result)}")
print('-' * 40)

# Print the first 10 results
pprint(result[0:10])
```

```python
# Convert the result to a Pandas DataFrame

hygiene0_df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
print(f"Number of rows in the data frame: {hygiene0_df.shape[0]}")

# Display the first 10 rows of the DataFrame
hygiene0_df.head(10)
```
